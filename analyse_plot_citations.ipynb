```python
import matplotlib as plt
import pandas as pd
dataCite_df = pd.read_csv('dataset_citation_counts_2000-01-01_to_2023-02-27_retrieved_28022023.csv')
crossRef_df = pd.read_csv('dataset_citation_publication_info_2000-01-01_to_2023-02-27_retrieved_28022023.csv')
```

```python
submitted = crossRef_df[crossRef_df['dataset_date_Type'] == 'Submitted']
submitted
```

```python
submitted.sort_values(by=['dataset_date'], ascending=False)
```

```python
submitted["dataset_date"] = submitted["dataset_date"].astype("datetime64")
```

```python
submitted.groupby(submitted["dataset_date"].dt.year).count().plot(kind="bar")
```
